```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
x = iris.data
y = iris.target
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)#1
print(iris)
```

```python
# - sepal length in cm\n
# - sepal width in cm\n
# - petal length in cm\n
# - petal width in cm\n
plt.figure(figsize=(8,8))
plt.scatter(x[:,0],x[:,1],c=y)#'setosa', 'versicolor', 'virginica'
plt.title("relations between Sepal Length and Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()#2
```

```python
pca = PCA(n_components=2)##42
x_pca = pca.fit_transform(x)
plt.figure(figsize= (8,8))
plt.scatter(x_pca[:, 0], x_pca[:, 1],c=y)
plt.title("after PCA")
plt.xlabel("n0")
plt.ylabel("n1")
plt.show()#3
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
j=knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
print("The accuracy is {}%".format(accuracy*100))#4
```


机器学习基础操作线性模型

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split# 划分测试集和训练集
from sklearn.linear_model import LinearRegression#线性回归模型
from sklearn.metrics import mean_squared_error#效果返回
from sklearn.preprocessing import StandardScaler#归一化
```

```python
df = pd.read_csv("bike.csv")#1
df = df.drop(columns="id")#2
df = df[df["city"] == 1]
df = df.drop(columns="city")#3
print(df)
print(df["hour"])
```

```python
df.loc[(df["hour"] >= 19) | (df["hour"] <= 5),"hour"] = 0
df.loc[(df["hour"] >= 6) & (df["hour"] <= 18),"hour"] = 1
print(df["hour"])#4
```

```python
arr_y = np.asarray(df["y"])
df = df.drop(columns="y")
print(df)
arr_y#5
```

```python
arr_df = np.asarray(df)
arr_df#6
```

```python
x_train,x_test,y_train,y_test = train_test_split(arr_df,arr_y,test_size=0.2,random_state=40)#7
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
y_scaler = StandardScaler()
y_train_scaled = y_scaler.fit_transform(y_train.reshape(-1,1))
y_test_scaled = y_scaler.transform(y_test.reshape(-1,1))#8
```

```python
model = LinearRegression()
model.fit(x_train_scaled,y_train_scaled)#9
```

```python
y_pred = model.predict(x_test_scaled)#10
```

```python
rmse = np.sqrt(mean_squared_error(y_test_scaled,y_pred))
print("RMSE=",rmse)#11
```
